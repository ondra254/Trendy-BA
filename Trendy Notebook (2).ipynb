```
import numpy as no
import pandas as pd
import matplotlib.pyplot as plt
import os 
from os import listdir

import cv2

dis_types = os.listdir('dataset')
print(dis_types)
print("Types of diseases found:", len(dis_types))
```

```
diseases = []
for item in dis_types
    all_diseases = os.listdir('dataset' + "/" +item)

for disease in all_diseases
    diseases.append((item, str('dataset' + "/" +item) + "/" + disease))
    print(diseases[:1])
```

```
disease_df = pd.DataFrame(data=diseases, columns=["disease type", "image"])
print(disease_df.head())
print(disease_df.tail())
```

```
print("Total number of diseases in dataset: ", len(disease_df))
```

```
disease_count = disease_df["disease type"].value_counts()
print("Examples in each category:" )
print(disease_count)
```
